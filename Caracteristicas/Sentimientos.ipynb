```python
# Importar las bibliotecas necesarias
import pandas as pd

from pysentimiento import create_analyzer
sentiment_analyzer = create_analyzer(task="sentiment", lang="es")

# Ruta al archivo CSV en tu Google Drive
#file_path = '/content/drive/MyDrive/Tesis- Borradores/FASE 1/TRAIN/data_train.csv'  # Cambia esto a la ruta real de tu archivo CSV
file_path = '/content/drive/MyDrive/Tesis- Borradores/Fase Final ()/Datos de Prueba/Datos/TRAIN Y TRIAL/trainytrial.csv'


# Leer el archivo CSV
df = pd.read_csv(file_path)

# Cargar el analizador de sentimientos
#sentiment_analyzer = SentimentAnalyzer(lang="es")

# Función para calcular el sentimiento
def calculate_sentiment(text):
    result = sentiment_analyzer.predict(text)
    sentiment = result.output  # Devuelve el sentimiento (POS, NEG, o NEU)
    pos_proba = result.probas['POS']  # Probabilidad de sentimiento positivo
    neg_proba = result.probas['NEG']  # Probabilidad de sentimiento negativo
    neu_proba = result.probas['NEU']  # Probabilidad de sentimiento neutral
    return sentiment, pos_proba, neg_proba, neu_proba

# Aplicar la función a la columna 'message' y crear las nuevas columnas 'sentiment', 'POS', 'NEG', 'NEU'
df[['sentiment', 'POS', 'NEG', 'NEU']] = df['message'].apply(lambda x: pd.Series(calculate_sentiment(x)))

# Guardar el DataFrame resultante en el mismo archivo CSV
df.to_csv(file_path, index=False)

print(f"Archivo procesado guardado en: {file_path}")
```
