```python
# Importar las bibliotecas necesarias
import pandas as pd
from transformers import pipeline

# Ruta al archivo CSV en tu Google Drive
file_path = 'linkCSV'

# Leer el archivo CSV
df = pd.read_csv(file_path)

# Cargar el modelo y el tokenizador
model_path = "citizenlab/distilbert-base-multilingual-cased-toxicity"
toxicity_classifier = pipeline("text-classification", model=model_path, tokenizer=model_path)

# Función para calcular la toxicidad
def calculate_toxicity(text):
    # Truncar el texto si excede el máximo de tokens permitidos (512)
    result = toxicity_classifier(text[:512])[0]  # Trunca a los primeros 512 caracteres
    # Devuelve la puntuación de toxicidad si el mensaje es tóxico, de lo contrario, devuelve 0
    toxicity_score = result['score'] if result['label'].lower() == 'toxic' else 0
    return toxicity_score

# Aplicar la función a la columna 'message' y crear la nueva columna 'toxicity'
df['toxicity'] = df['message'].apply(calculate_toxicity)

# Guardar el DataFrame resultante en el mismo archivo CSV
df.to_csv(file_path, index=False)

print(f"Archivo procesado guardado en: {file_path}")

```
